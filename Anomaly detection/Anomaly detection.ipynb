```python
def estimategaussian(X):
    m,n=X.shape
    mu=(1/m)*np.sum(X,axis=0)
    sigma=(1/(m-1))*np.sum(np.multiply((X-mu),(X-mu)),axis=0)
    
    return mu,sigma
```

```python
mu,sigma=estimategaussian(X)
```

```python
def multigaussian(X,mu,sigma):
    k=sigma.size
    if sigma.ndim==1:
        sigma=np.diag(sigma)
    X=X-mu
    p= (2*np.pi)**(-k/2) * (np.linalg.det(sigma)**(-0.5)*np.exp(-0.5*np.sum(np.dot(X,np.linalg.pinv(sigma))*X,axis=1)))  
    
    return p
```
