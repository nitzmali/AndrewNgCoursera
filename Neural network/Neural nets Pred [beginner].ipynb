```python
theta1=weights['Theta1']
theta2=weights['Theta2']
```

```python
nn_params = np.hstack((theta1.ravel(order='F'), theta2.ravel(order='F')))
```

```python
hidden_layer_size=25
input_layer_size=400
num_labels=10
```

```python
theta1 = np.reshape(nn_params[:hidden_layer_size*(input_layer_size+1)], (hidden_layer_size, input_layer_size+1), 'F')
theta2 = np.reshape(nn_params[hidden_layer_size*(input_layer_size+1):], (num_labels, hidden_layer_size+1), 'F')
```

```python
def sigmoid(X):
    return 1/(1+np.exp(-X))
```

```python
print(X.shape,y.shape,theta1.shape,theta2.shape)
```

```python
m=len(y)
ones=np.ones([m,1])
```

```python
a1=np.hstack((ones,X))
```

print(a1.shape,y.shape,theta1.shape,theta2.shape)

```python
print(a1.shape,y.shape,theta1.shape,theta2.shape)
```

```python
a2=sigmoid(a1@theta1.T)
a2=np.hstack((ones,a2))
```

```python
a3=sigmoid(a2@theta2.T)
```

```python
predicted=list(np.argmax(a3,axis=1)+1)
```

```python
newpred=[e if e else 10 for e in predicted]
```
