```python

#--------------------------------------------------------------------------------------------------------------------------
#Linear regrssion using Iteration one variable
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#profit=w1(population)+bias

#step1 predict function
def predict_sales(weight,population,bias):
    return weight*population+bias
    

#step 2 Cost function [Mean squared error MSE]
#goal is to minimize the squared error
#output is the single number representing the cost

'''MSE=1N∑i=1n(yi−(mxi+b))2'''

def cost_function(population,profit,weight,bias):
    n=len(population)
    total_error=0.0
    for i in range(n):
        total_error+=(profit[i]-((weight*population[i])+bias))**2
    
    return total_error/n

```

```python
#To Minimize the Cost function we use Gradient Descent
#Calculate the gradient descent of Cost Function

def update_weights(weight,bias,population,profit,learning_rate):
    n=len(population)
    weight_derivative=0
    bias_derivative=0
    for i in range(n):
        #Calculate partial derivatives
        weight_derivative+=-2*population[i]*(profit[i]-(population[i]*weight+bias))
        bias_derivative +=-2*(profit[i]-(population[i]*weight+bias))
        
    # We subtract because the derivatives point in direction of steepest ascent    
    weight-=(weight_derivative/n)*learning_rate
    bias-=(bias_derivative/n)*learning_rate
    
    return weight,bias
```

```python
#code Training our model
'''Training a model is the process of iteratively improving your prediction equation by 
looping through the dataset multiple times, each time updating the weight and bias values 
in the direction indicated by the slope of the cost function (gradient). Training is complete when we reach 
an acceptable error threshold, or when subsequent training iterations fail to reduce our cost.'''
#Before training we need to initialize our weights (set default values), set our hyperparameters 
#(learning rate and number of iterations), and prepare to log our progress over each iteration.

def train(population,profit,weight,bias,learning_rate,iters):
    cost_history=[]
    for i in range(iters):
        weight,bias= update_weights(weight,bias,population,profit,learning_rate)
        
        #Calculating the cost for auditing purpose
        cost_history.append(cost_function(population,profit,weight,bias))
        
        #Log progress
        
        if i%10 ==0:
            print("iter={:d}     weight={:.2f}  bias={:.4f} cost ={:.2f}".format(i,weight,bias,cost_history[i]))
    return weight,bias 
```

```python
with open('machine-learning-ex1\ex1\ex1data1.txt') as r:
    l=[each.strip().split(',') for each in r]
pop= [float(l[i][0]) for i in range(len(l))]
pro = [float(l[i][1]) for i in range(len(l))]
weight,bias=train(pop,pro,0.0,0.0,0.01,1000)
print(weight,bias)
print(predict_sales(weight,20.001,bias))
```

```python
#Linear regression using matrices
data =pd.read_csv('machine-learning-ex1\ex1\ex1data1.txt',header=None)
X=data.iloc[:,0] #read first column
y =data.iloc[:,1] #read second column
m=len(y)  #number of training example
data.head()

plt.scatter(X,y)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')#we add another dimesnion to the data 
plt.show()

#we add another dimension to the data to accomodate the intercept form

X=X[:,np.newaxis]
y=y[:,np.newaxis]
theta =np.zeros([2,1])
iterations =1500
alpha =0.01
ones=np.ones((m,1))
X=np.hstack((ones,X))
print(np.shape(X),np.shape(y))
```

```python
#Computing Cost
#divide by 2 to make derive calculations simple
def computecost(X,theta,y):
    temp=np.dot(X,theta)-y
    return np.sum(np.power(temp,2))/(2*m)
    
print(computecost(X,theta,y))
```

```python
#Gradient descent
def gradientdescent(X,y,theta,alpha,iterations):
    for _ in range(iterations):
        temp =np.dot(X,theta)-y
        temp=np.dot(X.T,temp)
        theta=theta-(alpha/m)*temp
    return theta
theta=gradientdescent(X,y,theta,alpha,iterations)
j=computecost(X,theta,y)
print(j)
print(theta)
```

```python
#Linear regression with multiple variables
#first column (size in square feet)
#second column (No of bedrooms)
#third column proces of house in dollars
import pandas as pd
data =pd.read_csv('machine-learning-ex1\ex1\ex1data2.txt',header=None)
#print(data)
X=data.iloc[:,0:2]
y=data.iloc[:,2:]
m=len(y)
data.head()
```

```python
'''By looking at the values, note that house sizes are about 1000 times the number of bedrooms. 
When features differ by orders of magnitude, 
first performing feature scaling can make gradient descent converge much more quickly.'''

'''Subtract the mean value of each feature from the dataset.
After subtracting the mean, additionally scale (divide) the feature values by their respective “standard deviations.”'''
normmean=np.mean(X)
normstd=np.std(X)
X=(X-normmean)/normstd
#print(X)
```

```python
#Adding the axis
#X=X[:,np.newaxis]
#y=y[:,np.newaxis]
ones=np.ones((m,1))
X=np.hstack((ones,X))
theta =np.zeros((3,1))
```

```python
#Compute cost

def computecost(theta,X,y):
    temp=np.dot(X,theta)-y
    return np.sum(np.power(temp,2))/(2*m)
    
print(computecost(theta,X,y))    
```

```python
def gradientdescent(theta,X,y,alpha,iters):
    m=len(y)
    for _ in range(iters):
        temp=np.dot(X,theta)-y
        temp=np.dot(X.T,temp)
        theta=theta-((alpha/m)*temp)
    return theta
```

```python

theta=gradientdescent(theta,X,y,0.01,2600)
print(theta)
print(computecost(theta,X,y))
```
